#### Pre-process & Lemmatization

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()
```

```python
import re
import string 

def preproc(sentence):
  text = sentence.lower()
  # text = text.decode('UTF-8')
  # hapus punc
  text = re.sub(r'[' + string.punctuation.translate("") + ']+', '', text)
  text = "<s> " + text + " </s>"
  text = text.replace('\n', ' </s> <s> ') 
  # hapus double space 
  text = re.sub(r'\s{2,}', ' ', text)
  text = lemmatizer.lemmatize(text)
  return text

text = preproc(file)
print(text)
```

##### Count the Probability

```python
def preduni(sentence):
  best = 0
  next = ''
  # random.shuffle(dictwords)
  
  for d in dictwords:
    sen = sentence + " " + d
    result = uni(sen)
    # print(result)
    if best < result:
      # print(sen)
      best = result
      next = d
  return next, best
```

```python
def predbi(sentence):
  best = 0
  next = ''
  # random.shuffle(dictwords)
  
  for d in dictwords:
    sen = sentence + " " + d
    result = bi(sen)
    # print(result)
    if best < result:
      # print(sen)
      best = result
      next = d
  return next, best
```

```python
def predtri(sentence):
  best = 0
  next = ''
  # random.shuffle(dictwords)
  
  for d in dictwords:
    sen = sentence + " " + d
    result = tri(sen)
    # print(result)
    if best < result:
      # print(sen)
      # print(result)
      best = result
      next = d
  return next, best
```

##### Comparison

```python
def predict(sentence):
  sentence = sentence.lower()
  uniword, uniprob = preduni(sentence)
  biword, biprob = predbi(sentence)
  triword, triprob = predtri(sentence)
  if uniprob > biprob and uniprob > triprob:
    print("Best = Unigram")
    print("Predict = {}".format(uniword))
    print("Final = {} {}".format(sentence, uniword))
    print("Probability = {}".format(uniprob))
    print()
    print("Bigram = {} {} {}".format(sentence, biword, biprob))
    print("Trigram = {} {} {}".format(sentence, triword, triprob ))
  if biprob > uniprob and biprob > triprob:
    print("Best = Bigram")
    print("Predict = {}".format(biword))
    print("Final = {} {}".format(sentence, biword))
    print("Probability = {}".format(biprob))
    print()
    print("Unigram = {} {} {}".format(sentence, uniword, uniprob ))
    print("Trigram = {} {} {}".format(sentence, triword, triprob))
  if triprob > biprob and triprob > uniprob:
    print("Best = Trigram")
    print("Predict = {}".format(triword))
    print("Final = {} {}".format(sentence, triword))
    print("Probability = {}".format(triprob))
    print()
    print("Unigram = {} {} {}".format(sentence, uniword, uniprob ))
    print("Bigram = {} {} {}".format(sentence, biword, biprob))
```
